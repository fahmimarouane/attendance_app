```python
import pandas as pd
import warnings
warnings.simplefilter("ignore", UserWarning)
pd.set_option('display.max_columns', None)
```

```python
# Load the Excel file as a DataFrame
file1 = 'export_notesCC_1APIC-3_0030.xlsx'
```

```python
df = pd.read_excel(file1, engine="openpyxl")
```

```python
# Extraire les colonnes 3 (index 2) à partir de la ligne 17 (index 16) pour le code massar
code_masssar = df.iloc[16:, 2].reset_index(drop=True)  # Colonne 3 (index 2), ligne 17 à la fin

# Extraire les colonnes 3 (index 2) à partir de la ligne 17 (index 16) pour les noms
nom = df.iloc[16:, 3].reset_index(drop=True)  # Colonne 3 (index 2), ligne 17 à la fin

# Créer un DataFrame avec ces deux colonnes
df_extracted = pd.DataFrame({
    'code_masssar': code_masssar,
    'nom': nom
})

df_extracted
```

```python
# Compter le nombre total d'élèves en utilisant la colonne 'code_masssar' (ou 'nom')
total_students = df_extracted['code_masssar'].notna().sum()
total_students
```

```python
df_annee = pd.read_excel(file1, skiprows=12, engine="openpyxl")
df_semestre = pd.read_excel(file1, skiprows=10, engine="openpyxl")
df_matiere = pd.read_excel(file1, skiprows=10, engine="openpyxl")
df_academie = pd.read_excel(file1, skiprows=6, engine="openpyxl")
df_province = pd.read_excel(file1, skiprows=6, engine="openpyxl")
df_ecole = pd.read_excel(file1, skiprows=6, engine="openpyxl")

df_niveau =  pd.read_excel(file1, skiprows=8, engine="openpyxl")
df_professeur  = pd.read_excel(file1, skiprows=8, engine="openpyxl")
df_classe =  pd.read_excel(file1, skiprows=8, engine="openpyxl")

# Extract information based on skiprows values
annee_scolaire = df_annee.columns[3]
semestre = df_semestre.columns[3]
matiere = df_matiere.columns[14]
academie = df_academie.columns[3]
province = df_province.columns[8]
ecole = df_ecole.columns[14]

niveau = df_niveau.columns[3]
professeur = df_professeur.columns[14]
classe = df_classe.columns[8]
```

```python
# Combine the extracted data into a dictionary
extracted_data = {
    'Annee Scolaire': annee_scolaire,
    'Semestre': semestre,
    'Matiere': matiere,
    'Academie': academie,
    'Province': province,
    'Ecole': ecole,
    'Niveau': niveau,
    'Professeur': professeur,
    'Classe': classe,
    
}

# Convert the dictionary into a DataFrame (optional)
df_extracted = pd.DataFrame([extracted_data])
df_extracted
```
